## Merge preprocessed datasets

```python
import pandas as pd
```

```python
final_node_categories = pd.read_csv('preprocessed_data/Product_NodeCategories.csv')
final_descriptions = pd.read_csv('preprocessed_data/Product_Descriptions.csv')
avg_df = pd.read_csv('preprocessed_data/AvgPriceAndSalesVolume.csv')
final_categories = pd.read_csv('preprocessed_data/ProductMainAndSubCategory.csv')
```

```python
final_node_categories.shape, final_node_categories.columns
```

```python
final_descriptions.shape, final_descriptions.columns
```

```python
avg_df.shape, avg_df.columns
```

```python
final_categories.shape, final_categories.columns
```

```python
m1 = pd.merge(final_categories, final_node_categories)

m2 = pd.merge(m1, final_descriptions)

m3 = pd.merge(m2, avg_df)
```

```python
m3
```

```python
m3['combined_string_full'] = m3['MAIN_CATEGORY'] + ". " + m3['SUB_CATEGORY'] + ". " + m3['MIN_CATEGORY_LEAF_NAME'] + ". " + m3['DESCRIPTION']
```

```python
m3['combined_string_categories'] = m3['MAIN_CATEGORY'] + ". " + m3['SUB_CATEGORY'] + ". " + m3['MIN_CATEGORY_LEAF_NAME']
```

```python
m3 = m3[['ASIN','MARKETPLACE','REGION','BRAND_NAME','combined_string_full','combined_string_categories',\
    'ORGANIC_UNITS_SOLD','SALE_PRICE_EUR']]
```

```python
m3
```

```python
m3.to_csv('preprocessed_data/Combined_Data.csv',index=False)
```
